# Other Groupings

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/dataset.xlsx")
df.head(5)
```

```python
conditions = [
    (df['Patient addmited to regular ward (1=yes, 0=no)'] == 1),
    (df['Patient addmited to semi-intensive unit (1=yes, 0=no)'] == 1),
    (df['Patient addmited to intensive care unit (1=yes, 0=no)'] == 1),
    (df['Patient addmited to intensive care unit (1=yes, 0=no)'] == 0) & 
     (df['Patient addmited to semi-intensive unit (1=yes, 0=no)'] == 0) &
     (df['Patient addmited to regular ward (1=yes, 0=no)'] == 0)]

# create a list of the values we want to assign for each condition
values = ['regular', 'semi', 'icu', 'discharged']

# create a new column and use np.select to assign values to it using our lists as arguments
df['y'] = np.select(conditions, values)

df['y'].value_counts()
```

```python
ids = df.loc[:, 'Patient ID':'Patient addmited to intensive care unit (1=yes, 0=no)']
col1 = df.loc[:,'Serum Glucose']
col2 = df.loc[:,'Urea':'Sodium']
col3 = df.loc[:,'Gamma-glutamyltransferase\xa0']
col4 = df.loc[:,'Alkaline phosphatase':'Magnesium']
col5 = df.loc[:,'Partial thromboplastin timeÂ (PTT)Â ':'Ferritin']
col6 = df.loc[:,'Lipase dosage':'Albumin']

other = pd.concat([ids, df['y'], col1, col2, col3, col4, col5, col6], axis=1)

# filter to only positive patients
other = other[other['SARS-Cov-2 exam result'] == 'positive']
other
```

`Albium`, `D-Dimer`, `Vitamin B12`, `Prothrombin time (PT), Activity` and `Partial thromboplastin time (PTT)` are missing for all covid positive patients so these variables will not be useful at all. In addition, if we use a 98% cut off for removing variables from our analysis, this would also include `Ferritin`, `Lipase dosage`, `Magnesium` and `Ionized calcium`.

Most of the other variables have different amounts of missing values meaning its likely they were not results of any grouping of lab tests done.

```python
other_missing = (other.isnull().sum().sort_values(ascending = False)).to_frame(name = 'num')
other_missing.reset_index(inplace=True)
other_missing.rename(columns = {'index':'var'}, inplace = True)
other_missing['prop'] = other_missing['num']/len(other)
other_missing[other_missing.prop > 0]
```

## Groupings by Labs

There don't seem to be any large groupings of test results in this other category, but there is a chance some of these variables fit in with other lab groupings.

The only grouping I know for sure within the other category is that `Sodium` and `Potassium` levels are tested together.

I want to check the number of missing values for other columns in the dataset to see if some of the columns in other fit in another grouping that we just didn't know to put together. In looking at the missing values for all other variables, I am not seeing any that seem to be in the same grouping as any of the 'other' variables.

```python
pd.set_option('display.max_rows', None)
covid_pos = df[df['SARS-Cov-2 exam result'] == 'positive']
missing_prop = (covid_pos.isnull().sum().sort_values(ascending = False)).to_frame(name = 'num')
missing_prop.reset_index(inplace=True)
missing_prop.rename(columns = {'index':'var'}, inplace = True)
missing_prop['prop'] = missing_prop['num']/len(covid_pos)
missing_prop[(missing_prop['prop'] <= 0.98) & (missing_prop['prop'] > 0)]
```

Since `Sodium` and `Potassium` are in a test together, I want to create a binary grouping for the two them.

```python
pd.set_option('display.max_rows', 25)
df['pot_sod_grp'] = np.where((df['Sodium'].notnull()) & (df['Potassium'].notnull()), 1, 0)
```

```python
df[['pot_sod_grp', 'y']].value_counts()
```
