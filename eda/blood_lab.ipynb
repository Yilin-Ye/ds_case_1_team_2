```python
data = pd.read_excel('C:/Users/yilin/Documents/dsi/ds_case_1_team_2/data/dataset.xlsx')
```

```python
conditions = [
    (data['Patient addmited to regular ward (1=yes, 0=no)'] == 1),
    (data['Patient addmited to semi-intensive unit (1=yes, 0=no)'] == 1),
    (data['Patient addmited to intensive care unit (1=yes, 0=no)'] == 1),
    (data['Patient addmited to intensive care unit (1=yes, 0=no)'] == 0) & 
     (data['Patient addmited to semi-intensive unit (1=yes, 0=no)'] == 0) &
     (data['Patient addmited to regular ward (1=yes, 0=no)'] == 0)]

# create a list of the values we want to assign for each condition
values = ['regular', 'semi', 'icu', 'discharged']

# create a new column and use np.select to assign values to it using our lists as arguments
data['y'] = np.select(conditions, values)

data['y'].value_counts()
```
